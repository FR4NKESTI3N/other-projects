```python
import numpy as np
import pandas as pd
from scipy.optimize import *

import matplotlib.pyplot as plt

%matplotlib inline

import seaborn as sns
import math
```

```python
train_data = pd.read_csv('train.csv')
# 28x28 images
n = 28
```

```python
input_node_count = n * n
layer1_node_count = 30
label_count = 10
```

```python
train_data['bias'] = np.ones(len(train_data))
```

```python
X_train = train_data.drop('label', axis = 1)[:100]
Y_train = pd.get_dummies(train_data['label'][:100])
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
def costFunction(theta, X, Y, lamb = 0):
    m = len(Y)
    
    cost = 0
    
    theta1 = theta[: ((input_node_count + 1) * layer1_node_count)].reshape(
                    (layer1_node_count, input_node_count + 1))
    theta2 = theta[((input_node_count + 1) * layer1_node_count) :].reshape(
                    (label_count, layer1_node_count + 1))
    
    m = len(X)
    
    for t in range(m):
        layer1_act = sigmoid(np.dot(theta1, X.loc[(t)]))
        layer1_act = np.append(1, layer1_act)
        result = sigmoid(np.dot(theta2, layer1_act))
    
        y = Y.loc[(t)]
        cost += (np.sum((y * np.log1p(result)) + ((1 - y) * np.log1p(1 - result))))/-m
    
    cost = cost
    print(cost)
    return cost
```

```python
Theta = np.random.rand((input_node_count + 1) * layer1_node_count + label_count * (layer1_node_count + 1))/10000
```

```python
costFunction(Theta, X_train, Y_train)
```

```python
def gradient(theta, X, Y, lamb = 0):
    m = len(Y)
    
    cost = 0
    
    theta1 = theta[: ((input_node_count + 1) * layer1_node_count)].reshape(
                    (layer1_node_count, input_node_count + 1))
    theta2 = theta[((input_node_count + 1) * layer1_node_count) :].reshape(
                    (label_count, layer1_node_count + 1))
    
    m = len(X)
    
    theta1_grad = np.zeros(theta1.shape)
    theta2_grad = np.zeros(theta2.shape)
    
    for t in range(m):
        layer1_act = sigmoid(np.dot(theta1, X.loc[(t)]))
        layer1_act = np.append(1, layer1_act)
        result = sigmoid(np.dot(theta2, layer1_act))
    
        y = Y.loc[(t)]
#         cost += np.sum((y * np.log1p(result)) + ((1 - y) * np.log1p(1 - result)))
        
        d3 = result - y
        d2 = np.dot(d3.T, theta2) * layer1_act * (1 - layer1_act)
        
#         print(d3.shape, d2.shape, layer1_act.shape)
#         d3 = d3.reshape((len(d3), 1))
        theta2_grad += np.dot(np.array(d3).reshape((len(d3), 1)), layer1_act.reshape((1, len(layer1_act))))
#         print(np.dot(theta1, X.loc[[t]].T))
        theta1_grad += np.dot(np.array(d2).reshape((len(d2), 1)), np.array(X.loc[(t)]).reshape((1, X.shape[1])))[1:]
#     cost = cost/-m
#     return cost
    theta1_grad /= m
    
    theta2_grad /= m
    
    theta1_grad = theta1_grad.reshape(((input_node_count + 1) * layer1_node_count, 1))
    theta2_grad = theta2_grad.reshape(((layer1_node_count + 1) * label_count, 1))
    
#     print(theta2_grad)
    
    grad = np.append(theta1_grad, theta2_grad)
    return grad
```
